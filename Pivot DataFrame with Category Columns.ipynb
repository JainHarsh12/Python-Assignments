```python

import pandas as pd

def create_unique_question_dataframe():
    """
    Creates a Pandas DataFrame with a unique question and its corresponding Python code answer.
    The question involves finding the second largest number in a list.

    Returns:
        pandas.DataFrame: DataFrame with 'Question' and 'Answer_Code' columns.
    """

    data = {'Question': ["Find the second largest number in a given list without using built-in functions like sorted() or nlargest()."],
            'Answer_Code': ["""
def find_second_largest(numbers):
    if len(numbers) < 2:
        return 'List must have at least two numbers'

    largest = float('-inf')  # Initialize to negative infinity
    second_largest = float('-inf')

    for number in numbers:
        if number > largest:
            second_largest = largest
            largest = number
        elif number > second_largest and number != largest:
            second_largest = number

    if second_largest == float('-inf'):
        return 'No second largest element found'

    return second_largest

# Example Usage:
my_list = [10, 5, 20, 8, 15, 20]
result = find_second_largest(my_list)
print(result)  # Output: 15

my_list = [1,1,1,1]
result = find_second_largest(my_list)
print(result) # Output: No second largest element found

my_list = [1]
result = find_second_largest(my_list)
print(result) #Output: List must have at least two numbers
"""]}
    df = pd.DataFrame(data)
    return df

# Example usage:
df_question = create_unique_question_dataframe()
df_question
```
